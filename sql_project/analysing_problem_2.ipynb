```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('top_paying_job_skills.csv')
data.head()
```

```python
# Count the occurrences of each skill
skill_counts = data['skills'].value_counts()
print(skill_counts)
```

```python
plt.style.use('dark_background')

plt.figure(figsize=(10, 6))

sns.barplot(x=skill_counts.values, y=skill_counts.index, palette='viridis')

plt.title('Distribution of Skills in Top-Paying Data Science Jobs', fontsize=16, color='white')
plt.xlabel('Number of Jobs Requiring Skill', fontsize=12, color='white')
plt.ylabel('Skills', fontsize=12, color='white')

# Add grid lines for readability (using dotted lines)
plt.grid(True, which='both', axis='x', linestyle='--', color='gray')

# Customize ticks color to be readable on the dark background
plt.xticks(color='white')
plt.yticks(color='white')

plt.tight_layout()
plt.show()
```
